```python
# Define initial weights, bias, learning rate and points for Adaline
eta = 0.25  # Learning rate

# Initial weights and bias
w1, w2, B = 0.1, 0.1, 0.1

# Points and their target values
data = [
    {'X1': 1, 'X2': 1,  'T': 1},    # I1
    {'X1': 1, 'X2': -1, 'T': -1},   # I2
    {'X1': -1, 'X2': 1, 'T': -1},   # I3
    {'X1': -1, 'X2': -1, 'T': 1},   # I4
    {'X1': 1, 'X2': 1, 'T': -1},    # I1 (repeat of a point)
    {'X1': 1, 'X2': -1, 'T': 1},    # I2 (repeat)
    {'X1': -1, 'X2': 1, 'T': 1},    # I3 (repeat)
    {'X1': -1, 'X2': -1, 'T': -1}   # I4 (repeat)
]

# List to store the outputs
results = []

# Function to calculate Adaline updates
def adaline_update(w1, w2, B, X1, X2, T, eta):
    # Predict Y_in
    Y_in = X1 * w1 + X2 * w2 + B
    
    # Calculate the error
    error = T - Y_in
    
    # Update weights and bias
    w1_new = w1 + eta * error * X1
    w2_new = w2 + eta * error * X2
    B_new = B + eta * error
    
    # Return the updated weights, bias, and the error
    return w1_new, w2_new, B_new, Y_in, error

# Apply Adaline for each point
for point in data:
    X1, X2, T = point['X1'], point['X2'], point['T']
    
    # Get updated weights, bias, prediction and error
    w1, w2, B, Y_in, error = adaline_update(w1, w2, B, X1, X2, T, eta)
    
    # Store the results for each iteration
    results.append({
        'X1': X1, 'X2': X2, 'T': T, 'Y_in': Y_in, 'Error': error,
        'Updated w1': w1, 'Updated w2': w2, 'Updated B': B
    })

results
```
