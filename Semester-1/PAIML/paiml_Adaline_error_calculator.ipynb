```python
arr = [[1, 1], [1, -1], [-1, 1], [-1, -1]]
t = [1, -1, -1, -1]
w1 = 0.1
w2 = 0.1
b = 0.1

for i in range(100):
    avg = 0
    print("EPOCH", i + 1, "Errors\n")
    for j in range(4):
        
        yi = arr[j][0] * w1 + arr[j][1] * w2 + 1 * b
        dif = t[j] - yi


        w1 += 0.25 * dif * arr[j][0]
        w2 += 0.25 * dif * arr[j][1]
        b += 0.25 * dif * 1
        err = dif * dif
        print(err, end=" ")
        avg += err
    print(f"\n\nTotal Mean Error : {avg}\n\n")
```
